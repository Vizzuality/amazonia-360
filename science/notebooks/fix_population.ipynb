```python
import numpy as np
import rasterio as rio
```

```python
with rio.open("../data/raw/AFP GGHS_POP_E2025_R2023A 180 m WGS84.tif") as src:
    data = src.read(1)
    profile = src.profile.copy()
    old_nodata = src.nodata
print(profile)
```

```python
profile.update({"compress": "zstd", "nodata": -1, "dtype": "float32"})
profile.update({"nodata": -1})

with rio.open("../data/processed/population_fix.tif", "w", **profile) as dst:
    dst.write(np.where(data == old_nodata, -1, data.astype(np.float32)), 1)
```
